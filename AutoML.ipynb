```python
#Importing necessary libraries
import h2o
from h2o.automl import H2OAutoML
```

```python
#Starting the H2O cluster
h2o.init()
```

```python
# Loading the dataset
file_path = "/content/drive/MyDrive/4th SEM/ML/Assignment /diabetes.csv"
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
data = h2o.import_file(file_path, col_names=names)
```

```python
#Converting Outcome to categorical
data['Outcome'] = data['Outcome'].asfactor()
```

```python
#Splitting features and target
X = data.drop('Outcome', axis=1)
y = data['Outcome']
```

```python
#Spliting data into train and test sets
train, test = data.split_frame(ratios=[0.8])
```

```python
# Training AutoML
aml = H2OAutoML(max_runtime_secs=120)
aml.train(y='Outcome', training_frame=train)
```

```python
#Getting the best model
best_model = aml.leader
```

```python
#Performance metrics
performance = best_model.model_performance(test_data=test)
print(performance)
```

```python
#Making predictions on test data
predictions = best_model.predict(test)
```

```python
#Converting H2O frame to pandas dataframe
y_true = test['Outcome'].as_data_frame()
y_pred = predictions.as_data_frame()['predict']

#Performance metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print("Accuracy:", accuracy_score(y_true, y_pred))
print("Precision:", precision_score(y_true, y_pred))
print("Recall Score:", recall_score(y_true, y_pred))
print("F1 Score:", f1_score(y_true, y_pred))
```

```python
# Shutting down H2O cluster
h2o.cluster().shutdown()
```

```python
#Importing necessary libraries
import pandas as pd
```

```python
#Loading the dataset
data = pd.read_csv("/content/drive/MyDrive/4th SEM/ML/Assignment /diabetes.csv")
```

```python
#Splitting features and target
X = data.drop('Outcome', axis=1)
y = data['Outcome']
```

```python
#Splitting the data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
#Initializing the TPOT classifier
from tpot import TPOTClassifier
tpot = TPOTClassifier(generations=5, population_size=20, verbosity=2, random_state=7, n_jobs=-1, max_time_mins=5)
tpot.fit(X_train, y_train)
```

```python
#Evaluating the model
y_pred = tpot.predict(X_test)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall Score:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```
